```python
from PIL import Image
```

# Resizing the image

```python
image= Image.open('dog1.jpeg')
# image.show()
new_image=image.resize((200,200))
new_image.save('edited_dog.jpeg')
new_image.show()
```

# convert color image into greyscale

```python
img= image.convert('L')
img.save('grey_dog.jpeg')
img.show()
```

# convert grey image to color image

```python
gray_img= Image.open('grey_dog.jpeg')
color_img= Image.merge('RGB', [gray_img]*3)
color_img.save('color_img.jpeg')
color_img.show()
```

# convert color/grey image to black & white

```python
black_and_white_img= image.convert('1')
black_and_white_img.save('blackwhite_dog.jpeg')
black_and_white_img.show()
```

# seperate color image into r g and b planes

```python
import cv2
import numpy as np

img = cv2.imread('dog1.jpeg')

blueimg, redimg, greenimg= cv2.split(img)
zeros= np.zeros(blueimg.shape, np.uint8)

blue_bg_img= cv2.merge([blueimg,zeros,zeros])
green_bg_img= cv2.merge([zeros,greenimg,zeros])
red_bg_img= cv2.merge([zeros,zeros,redimg])



cv2.imshow("red",red_bg_img)
cv2.waitKey(0)
cv2.imshow("green",blue_bg_img)
cv2.waitKey(0)
cv2.imshow("blue",green_bg_img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

# create color image using three seperate R, G, B planes

```python
color_image = cv2.merge([blueimg, greenimg, redimg])
cv2.imshow("color",color_image)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

# Create image profile

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Read the image
img = cv2.imread('dog1.jpeg')

# Split the image into color channels
blue_img, green_img, red_img = cv2.split(img)

# Display the original image
plt.figure(figsize=(10, 5))
plt.subplot(2, 3, 1)
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
plt.title('Original Image')

# Display the individual color channels
plt.subplot(2, 3, 2)
plt.imshow(blue_img, cmap='Blues')
plt.title('Blue Channel')

plt.subplot(2, 3, 3)
plt.imshow(green_img, cmap='Greens')
plt.title('Green Channel')

plt.subplot(2, 3, 4)
plt.imshow(red_img, cmap='Reds')
plt.title('Red Channel')

# Display the histograms for each color channel
plt.subplot(2, 3, 5)
plt.hist(blue_img.flatten(), bins=256, color='blue', alpha=0.7)
plt.title('Blue Histogram')

plt.subplot(2, 3, 6)
plt.hist(green_img.flatten(), bins=256, color='green', alpha=0.7)
plt.title('Green Histogram')

plt.subplot(2, 4, 7)
plt.hist(red_img.flatten(), bins=256, color='red', alpha=0.7)
plt.title('Red Histogram')

plt.show()
```
